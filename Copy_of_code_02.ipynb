# 1a parte

No espaço de código abaixo, plote as 3 imagens, uma ao lado da outra. Ainda, plote o histograma de cada canal de cor (R - Vermelho, G - Verde e B - Azul) abaixo da respectiva imagem.

```
imagem1 = Image.open('/content/drive/My Drive/Faculdade/5 Período/Computação Gráfica/exercicio 02 04/bailey-zindel-NRQV-hBF10M-unsplash.jpg')
imagem2 = Image.open('/content/drive/My Drive/Faculdade/5 Período/Computação Gráfica/exercicio 02 04/claudio-testa--SO3JtE3gZo-unsplash.jpg')
imagem3 = Image.open('/content/drive/My Drive/Faculdade/5 Período/Computação Gráfica/exercicio 02 04/johannes-plenio-RwHv7LgeC7s-unsplash.jpg')

plt.figure(figsize=(17,4))

plt.subplot(131) #131 1 linha 3 colunas posição 1
plt.title('Imagem 1')
plt.axis('off')
plt.imshow(imagem1)

plt.subplot(132) #131 1 linha 3 colunas posição 1
plt.title('Imagem 2')
plt.axis('off')
plt.imshow(imagem2)

plt.subplot(133) #131 1 linha 3 colunas posição 1
plt.title('Imagem 3')
plt.axis('off')
plt.imshow(imagem3)



imagem1_arr = np.array(imagem1)
imagem2_arr = np.array(imagem2)
imagem3_arr = np.array(imagem3)

# vermelho

red1 = imagem1_arr[:,:,0]
red2 = imagem2_arr[:,:,0]
red3 = imagem3_arr[:,:,0]

plt.figure(figsize=(17,4))

plt.subplot(231) #131 1 linha 3 colunas posição 1
plt.hist(red1.ravel(), color='red')

plt.subplot(232) #131 1 linha 3 colunas posição 1
plt.hist(red2.ravel(), color='red')

plt.subplot(233) #131 1 linha 3 colunas posição 1
plt.hist(red3.ravel(), color='red')

green1 = imagem1_arr[:,:,1]
green2 = imagem2_arr[:,:,1]
green3 = imagem3_arr[:,:,1]

# verde
plt.figure(figsize=(17,4))

plt.subplot(231) #132 1 linha 3 colunas posição 2
plt.hist(green2.ravel(), color='green')

plt.subplot(232) #132 1 linha 3 colunas posição 2
plt.hist(green2.ravel(), color='green')

plt.subplot(233) #132 1 linha 3 colunas posição 2
plt.hist(green3.ravel(), color='green')


blue1 = imagem1_arr[:,:,2]
blue2 = imagem2_arr[:,:,2]
blue3 = imagem3_arr[:,:,2]

# azul
plt.figure(figsize=(17,4))

plt.subplot(231) #133 1 linha 3 colunas posição 3
plt.hist(blue1.ravel(), color='blue')

plt.subplot(232) #133 1 linha 3 colunas posição 3
plt.hist(blue2.ravel(), color='blue')

plt.subplot(233) #133 1 linha 3 colunas posição 3
plt.hist(blue3.ravel(), color='blue')
```

Olhando para os gráficos gerados, discorram comparando os resultados.

R: Na primeira imagem, a cor vermelha tem bastante nos tons mais fortes e nos tons mais fracos, ambos perto de 6, ja o verde, tem mais abundancia nos tons intermediarios, indo ate 3 apenas, no azul e a cor q mais tem na imagem, indo ate 7.5 e ainda nos tons mais forte.
  Na segunda imagem, as cores vermelho e verde sao bem proporcional em tons e quantidade das cores, no azul tambem tem a mesma abundancia de cor, mas o azul dela é mais fraco.
  Na terceira imagem, a cor mais forte é o vermelho, mas ela nao esta em maior abundancia, a cor que mais tem abundacia é o azul e tem o dobro da cor vermelha, mas o tom do azul é o mais fraco, assim so mostrando a cor vermelha na imagem.

Escolha uma das imagens acima, para continuarmos os experimentos. 

Vamos gerar algumas imagens preto e branco, utilizando limiares diferentes:

*  45
*  100
*  145
*  200
*  245

Escreva o código no espaço abaixo.

```

img45 = np.mean(imagem2_arr, axis = 2)
img45[img45 >= 45] = 255
img45[img45 < 45] = 0
#plt.subplot(151)
#plt.imshow(Image.fromarray(img45))

img100 = np.mean(imagem2_arr, axis = 2)
img100[img100 >= 100] = 255
img100[img100 < 100] = 0
#plt.subplot(152)
#plt.imshow(Image.fromarray(img100))

img145 = np.mean(imagem2_arr, axis = 2)
img145[img145 >= 145] = 255
img145[img145 < 145] = 0
#plt.subplot(153)
#plt.imshow(Image.fromarray(img145))

img200 = np.mean(imagem2_arr, axis = 2)
img200[img200 >= 200] = 255
img200[img200 < 200] = 0
#plt.subplot(154)
#plt.imshow(Image.fromarray(img200))

img245 = np.mean(imagem2_arr, axis = 2)
img245[img245 >= 245] = 255
img245[img245 < 245] = 0
#plt.subplot(155)
#plt.imshow(Image.fromarray(img245))
```

Plote as imagens geradas. (Escreva o código no espaço abaixo)

```
plt.figure(figsize=(24,4))

plt.subplot(151)
plt.imshow(Image.fromarray(img45))

plt.subplot(152)
plt.imshow(Image.fromarray(img100))

plt.subplot(153)
plt.imshow(Image.fromarray(img145))

plt.subplot(154)
plt.imshow(Image.fromarray(img200))

plt.subplot(155)
plt.imshow(Image.fromarray(img245))
```
